```python
import json
from pprint import pprint
```

```python
PATH_TO_SUMMARIES = '../../data/patients/patients_1.0_summaries.json'
PATH_TO_TRANSCRIPTS = '../../data/patients/patients_1.0_with_transcripts.json'
```

```python
def read_json(path):
    with open(path, 'r') as f:
        return json.load(f)
```

```python
summaries = read_json(PATH_TO_SUMMARIES)
```

```python
transcripts = read_json(PATH_TO_TRANSCRIPTS)
```

```python
transcripts
```

```python
def filter_and_pprint(id):
    """Filter to specified id in the file dictionary and pretty print"""
    print("TRANSCRIPTS FOR ID: ", id)
    pprint(transcripts[id]['chat_transcript'], indent=2, width=150)
    print()
    print("SUMMARIES FOR ID: ", id)
    pprint(summaries[id]['summary'], indent=2, width=150)

filter_and_pprint('14185111')
```
