```python
import json

def get_patient_details(data, id, name):
    # Check if the ID exists in the data
    if str(id) in data:
        # Check if the name matches
        if data[str(id)]["name"] == name:
            return data[str(id)]["prompt"], data[str(id)]["chat_transcript"]
        else:
            return "Name does not match the provided ID.", None
    else:
        return "ID not found.", None

def load_json_file(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)

# Path to the JSON file
json_file_path = "/Users/faridgholitabar/Coding/reco/reco_analysis/data/patients/patients_1.0_with_transcripts.json"

# Load JSON data into a Python dictionary
data = load_json_file(json_file_path)

# Example usage
id_input = 13912736
name_input = "Jennifer Collins"
prompt, transcript = get_patient_details(data, id_input, name_input)

if transcript is not None:
    print("Prompt:")
    print(prompt)
    print("\nChat Transcript:")
    for line in transcript:
        print(line)
else:
    print(prompt)
```
